The forklifts can only access a roll of paper if there are fewer than four rolls of paper in the eight adjacent positions. If you can figure out which rolls of paper the forklifts can access, they'll spend less time looking and more time breaking down the wall to the cafeteria.

```python
import numpy as np
```

```python
def count_adjacent(data):
    x = len(data[0])
    y = len(data)
    accessible = data.copy()
    counts = 0
    for row in range(0, x):
        #print("row ", row)
        for col in range(0, y):
            #print(data[row][col])
            if data[row][col] == '@':#print(row, col)
                row_start = max(row-1, 0)
                row_end = min(row+1, x-1)
                col_start = max(col-1, 0)
                col_end = min(col+1, y)
                #print(row_start, row_end, col_start, col_end)
                #print(data[row_start:row_end+1, col_start:col_end+1])
                adj = count_ats(data[row_start:row_end+1, col_start:col_end+1].tolist())
                #print("col ", col, "adj", adj)                
                if adj<=4:
                    accessible[row][col] = "x"
                    counts +=1
    return counts, accessible
    
def count_ats(adj):
    #print(adj)
    count = 0
    for element in adj:
        count += element.count("@")
    #print(count)
    return count
            
print(test_data)
test_counts, test_accessible = count_adjacent(test_data)
print(test_counts)
```
