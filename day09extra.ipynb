```python
elf_data = []

with open('day09.txt', 'r') as file:
    for line in file:
        elf_data.append(line.replace("\n", ""))

ed = []
for i in range(len(elf_data)):
    ed.append(elf_data[i].split(','))
    elf_data[i] = [int(datum) for datum in ed[i]]
    
print(len(elf_data))
print(elf_data[0:10])
print(type(elf_data[0][0]))
```

```python
test_data = []

with open('test09.txt', 'r') as file:
    for line in file:
        test_data.append(line.replace("\n", ""))

td = []
for i in range(len(test_data)):
    td.append(test_data[i].split(','))
    test_data[i] = [int(datum) for datum in td[i]]

print(len(test_data))
print(test_data)
print(type(test_data[0][0]))

# https://www.reddit.com/r/adventofcode/comments/1pi5rqn/2025_day_9_part_2_check_your_solution_with_this/
```

```python
def plotting(data):
    x_c = [x for (x, y) in data]
    y_c = [y for (x, y) in data]

    max_x = max(x_c)+1
    min_x = min(x_c)-1
    max_y = max(y_c)+1
    min_y = min(y_c)-1

    print(max_x, min_x, max_y, min_y)

    grid = []
    for y in range (min_y, max_y+1):
        row=[]
        for x in range (min_x, max_x+1):
            if [x, y] in data:
                row.append("#")
            else:
                row.append(".") 
        grid.append(row)
        
    return grid


def print_grid(grid):
    for g in grid:
        print(g)


test_grid = plotting(test_data) 
print_grid(test_grid)
    
```

```python
elf_grid = plotting(elf_data)

print_grid(elf_grid)
```
