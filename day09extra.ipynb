```python
def plotting(data):
    x_c = [x for (x, y) in data]
    y_c = [y for (x, y) in data]

    max_x = max(x_c)+1
    min_x = min(x_c)-1
    max_y = max(y_c)+1
    min_y = min(y_c)-1

    print(max_x, min_x, max_y, min_y)

    grid = []
    for y in range (min_y, max_y+1):
        row=[]
        for x in range (min_x, max_x+1):
            if [x, y] in data:
                row.append("#")
            else:
                row.append(".") 
        grid.append(row)
        
    return grid


def print_grid(grid):
    for g in grid:
        print(g)


test_grid = plotting(test_data) 
print_grid(test_grid)
    
```
